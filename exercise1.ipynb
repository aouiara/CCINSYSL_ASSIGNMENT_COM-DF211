2. Find the function **f(x)**  

| Input  | CAT | DOG | APPLE | BANANA | ORANGE |  
| --- | --- | --- | --- | --- | --- |
| Output | CT | DG | PPL | BNN | RNG


```python
def antiVowel(Input):
    vowels = 'aeiouAEIOU'
    consonants = ""

    for char in Input:
      if char not in vowels:
        consonants += char
    return consonants
```
